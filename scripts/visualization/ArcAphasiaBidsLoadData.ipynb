```python
from datasets import load_dataset

ds = load_dataset("hugging-science/arc-aphasia-bids", streaming=True)
```

```python
example = None
for sample in ds['train']:
    if sample['t2w'] is not None and sample['lesion'] is not None:
        example = sample
        break
```

```python
import pickle
from pathlib import Path

# Create directory if it doesn't exist
output_dir = Path("data/file_with_t2w_lesion")
output_dir.mkdir(parents=True, exist_ok=True)

# Save the example
output_file = output_dir / "example.pkl"
with open(output_file, "wb") as f:
    pickle.dump(example, f)

print(f"Saved example to {output_file}")
```
