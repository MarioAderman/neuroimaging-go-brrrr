```python
import pickle

with open("data/file_with_t2w_lesion/example.pkl", "rb") as f:
    example = pickle.load(f)
```

```python
from ipyniivue import NiiVue, ShowRender, SliceType, Volume
from IPython.display import display

nv = NiiVue()
nv.set_slice_type(SliceType.MULTIPLANAR)
nv.opts.multiplanar_show_render = ShowRender.ALWAYS
nv.opts.show_3d_crosshair = True
nv.opts.multiplanar_force_render = True
```

```python
from ipyniivue import NiiVue, ShowRender, SliceType, Volume
from IPython.display import display

if example:
    nv = NiiVue()
    nv.set_slice_type(SliceType.MULTIPLANAR)
    nv.opts.multiplanar_show_render = ShowRender.ALWAYS
    nv.opts.show_3d_crosshair = True
    nv.opts.multiplanar_force_render = True
    
    # Create volumes: ORDER MATTERS - base layer last, overlays first
    volumes = [
        Volume(
            name="t2w.nii.gz",
            data=example['t2w'].to_bytes(),
            colormap='gray',
            opacity=1.0
        ),
        Volume(
            name="lesion.nii.gz",
            data=example['lesion'].to_bytes(),
            colormap='red',
            opacity=0.5
        ),
    ]
    
    # Load and display
    nv.load_volumes(volumes)
    display(nv)
    
    print(f"Displaying subject: {example['subject_id']}, session: {example['session_id']}")
    print(f"Age at stroke: {example['age_at_stroke']}, Sex: {example['sex']}")
    print(f"WAB AQ: {example['wab_aq']}, Type: {example['wab_type']}")
else:
    print("No samples found with both t2w and lesion data")
```
